# **4. Começando wordcloud**

Como sabemos provavelmente alguns data sets apresentam ausência de dados, e isto possivelmente poderá resultar em algum erro na execução worcloud.

O primeiro comando que deve ser executado é a remoção das linhas onde apresentam dados ausentes na coluna summary, nesta variável encontra-se os textos utilizados para realizar o wordcloud.

```
df.dropna(subset=['summary'], axis=0, inplace = True)
```

Foi utilizado a função dropna para a remoção das linhas, essa função é utilizada para apagar as linhas que não possuem dados de uma respectiva coluna, no caso a coluna summary.

Após ser apagado as linhas, foi atribuído a coluna summary para uma variável chamada summay.

```
summary = df['summary']
```

Após ser apagado as linhas, foi atribuído a coluna summary para uma variável chamada summay, e essa variável será utilizada daqui em diante.

Mostrando a variável summary.

```
summary
```

Para a realização de um wordcloud em python é necessário que todas as palavras estejam juntas, todas contidas em apenas uma string.

A partir disto, abaixo foi utilizado a função join, onde está irá colocar todas as linhas em somente uma string, e separadas por um espaço "  ", como mostrado no código abaixo.

```
all_summary = " ".join(s for s in summary)
```

Nota-se que toda junção da coluna foi atribuída a variável all_summary, assim foi deixada todas as descrições em um somente uma string, sendo esta praticamente um texto, está quase tudo pronto para criação do wordcloud.

A biblioteca STOPWORDS do pacote wordcloud é utilizada para eliminar provérbios e algumas palavras que devem ser excluídas do texto, exemplo (de, do, da, em, na, no, I, you) e entre outras.
Abaixo encontra-se o código com as stopwords, esta que foi atribuída a variável stopwords.

```
stopwords = set(STOPWORDS) 
```

Ainda é preciso realizar mais uma etapa para stopwords, pois, existem algumas palavras que não contidas neste conjunto da biblioteca, logo abaixo será atribuído variável stopword as principais palavras não contidas no conjunto.

Algumas preposições que não estão no conjunto são ("da", "meu", "em", "você", "de", "ao", "os").

```
stopwords.update(["da", "meu", "em", "você", "de", "ao", "os"])
```

Agora chegou a etapa final da criação do wordcloud.

Primeiramente será usada a função WordCloud, nesta foi colocado alguns argumentos, como as stopwords, definido a cor do fundo da imagem, e as dimensões da imagem.

Por fim é utilizado o método generate e atribuir a variável all_summary que contém o texto que irá gerar o wordcloud.

```
wordcloud = WordCloud(stopwords=stopwords,
                      background_color='black', width=1600, height=800).generate(all_summary)
```

Tudo isso foi atribuído a variável wordcloud, e agora é necessário gerar uma imagem a partir deste.

Abaixo foi criado um subplot e utilizada a função imshow para gerar a imagem do wordcloud.

Após foi utilizada a função to_file para gerar um arquivo png para download, para realização do download deve-se ir na pasta files do google colaboratory.

```


fig, ax = plt.subplots(figsize=(16,8))
ax.imshow(wordcloud, interpolation='bilinear')
ax.set_axis_off()

plt.imshow(wordcloud)
wordcloud.to_file('rafael.png',);

```
